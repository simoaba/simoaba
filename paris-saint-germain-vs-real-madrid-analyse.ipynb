#### **<span style="color:blue">Detailed</span>**

```r
library(tidyverse)
library(hrbrthemes)
library(readxl)
```

```r
stats <- read_excel("../input/rmvspsg/RMvsPSG.xlsx", sheet = "accurate")
stats_lon <- pivot_longer(stats, cols = 2:18, names_to ="name", values_to = "per game")
head(stats)
```

**The data collected in 1 table with the following stats per game in percentage of accuracy: **  
* Cross	
* Corner	
* Long Balls	
* Short Passes	
* Aerial Duels	
* Tackles Won	
* Dribbled Past   
* Out Of Box Shots	
* Penalty Area Shots	
* Open Play Shots	
* Set Piece Shots	
* Penalty Area Goal	    
* Out Of Box Goal	
* Open Play Goal	
* Set Piece Goal	
* Unsuccessful Dribbles	    
* Successful Dribbles	

```r
stats_lon %>%
  filter(name %in% c("Cross","Corner","Aerial Duels","Long Balls","Short Passes"))%>%
  ggplot(aes(Team, name, fill= `per game`)) + 
  geom_tile()+
  geom_text(aes(Team, name, label = `per game`), color = "black", size = 4)+
  scale_fill_gradient2(low = "blue", high = "red", mid = "white",midpoint = 50, limit = c(0,100), space = "Lab",
                       name="Accuracy per game\n(%)") +
  theme_ipsum()+
  theme(axis.title.x=element_blank(), axis.title.y=element_blank())
```

*** We notice that both teams plays well the short passes with a high accuracy (more than 90% for both). But Real Madrid has a high long balls accuracy, Cross and Aerial duels. But PSG rally on the qualities that Sergio Ramos brings to can face and beats the odds.**

```r
stats_lon %>%
  filter(name %in% c("Tackles Won","Dribbled Past","Unsuccessful Dribbles","Successful Dribbles"))%>%
  ggplot(aes(Team, name, fill= `per game`)) + 
  geom_tile()+
  geom_text(aes(Team, name, label = `per game`), color = "black", size = 4)+
  scale_fill_gradient2(low = "blue", high = "red", mid = "white",midpoint = 50, limit = c(0,100), space = "Lab",
                       name="Accuracy per game\n(%)") +
  scale_y_discrete(limits = c("Tackles Won","Dribbled Past","Unsuccessful Dribbles","Successful Dribbles"))+
  theme_ipsum()+
  theme(axis.title.x=element_blank(), axis.title.y=element_blank())
```

*** PSG with a talented player as Messi, Neymar and mbappé will be faced the qualities of Real Madrid's defender. since only 32.4% of dribbles past. but always the individual work can e the key in complicated games like this in certain games.**

```r
stats_lon %>%
  filter(name %in% c("Penalty Area Goal","Out Of Box Goal","Open Play Goal","Set Piece Goal",
                     "Out Of Box Shots","Penalty Area Shots","Open Play Shots","Set Piece Shots"))%>%
  ggplot(aes(Team, name, fill= `per game`)) + 
  geom_tile()+
  geom_text(aes(Team, name, label = `per game`), color = "black", size = 4)+
  scale_fill_gradient2(low = "blue", high = "red", mid = "white",midpoint = 50, limit = c(0,100), space = "Lab",
                       name="Accuracy per game\n(%)") +
  scale_y_discrete(limits = c("Penalty Area Goal","Out Of Box Goal","Penalty Area Shots","Out Of Box Shots",
                              "Open Play Goal","Set Piece Goal","Open Play Shots","Set Piece Shots"))+
  theme_ipsum()+
  theme(axis.title.x=element_blank(), axis.title.y=element_blank())
```

*** In the attacking style, PSG most of the danger come from Open Play game and inside the penalty Area. and the inverse for Real Madrid that there's a balance in the attack between the shooting from inside the box and outside the box. And also 21% of the shots and the goal come from the Set Piece.**
