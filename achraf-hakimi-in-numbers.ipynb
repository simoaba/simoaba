#### **<span style="color:blue">Share</span>**
##### **Hakimi's performance in 4 years**

```r
h1 <- hakimi %>%
  ggplot( aes(x=Season, y=Clearances,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h2 <- hakimi %>%
  ggplot( aes(x=Season, y=ShotsBlocked,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h3 <- hakimi %>%
  ggplot( aes(x=Season, y=CrossesBlocked,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h4 <- hakimi %>%
  ggplot( aes(x=Season, y=PassesBlocked,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h5 <- hakimi %>%
  ggplot( aes(x=Season, y=TotalTackles,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h6 <- hakimi %>%
  ggplot( aes(x=Season, y=DribbledPast,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h7 <- hakimi %>%
  ggplot( aes(x=Season, y=Interceptions,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h8 <- hakimi %>%
  ggplot( aes(x=Season, y=FoulsCommited,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
h9 <- hakimi %>%
  ggplot( aes(x=Season, y=DualsWon,group = 1,fill=Season)) +
  geom_col(width=0.5)+
  geom_line()+
  geom_point()+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
hakimi_graph<-(h1/h2/h3)|(h4/h5/h6)|(h7/h8/h9)

hakimi_graph + plot_layout(guides = 'collect')+
  plot_annotation(title = "Achraf Hakimi stats per 90mins",subtitle="From 2018/2019 to 2021/2022")&
  theme(legend.position = "top")
```

##### **Summary**
After 24 games for Hakimi with the PSG this year, we can summarize the defensive level of Hakimi this year in 2 points.  
**Positive observation:**   
* This year Hakimi committed less fouls (with 0.6 fouls per 90mins). This is a good point for Hakimi since his history with yellow cards is not so good.
* The Clearances number (1.25 per 90mins), he clears more balls in defensive third, and this can be the reason of the change of the formation from 3-5-2 to 4 in the back with Pochettino and this makes him more involved in the defensive.

**The negative point this year:**   
* This year Hakimi plays with the formation of 4 in back but his number is not doing good, as a fullback he is the fourth player in the Ligue 1 that he gets dribbled 1.4 per game.
* There is also a decrease for the other defensive qualities.

We can say that Hakimi is having a bad year this season; nevertheless, this can be normal as he joins the PSG team with a different playing conditions and new tactics, but he can still improve his numbers in the second half of the season.

##### **The comparison of the full back performance**

```r
f1 <- fullback %>%
  ggplot( aes(x=Player, y=Clearances,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f2 <- fullback %>%
  ggplot( aes(x=Player, y=ShotsBlocked,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f3 <- fullback %>%
  ggplot( aes(x=Player, y=CrossesBlocked,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f4 <- fullback %>%
  ggplot( aes(x=Player, y=PassesBlocked,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f5 <- fullback %>%
  ggplot( aes(x=Player, y=TotalTackles,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f6 <- fullback %>%
  ggplot( aes(x=Player, y=DribbledPast,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f7 <- fullback %>%
  ggplot( aes(x=Player, y=Interceptions,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f8 <- fullback %>%
  ggplot( aes(x=Player, y=FoulsCommited,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
f9 <- fullback %>%
  ggplot( aes(x=Player, y=DualsWon,group = 1,fill=Player)) +
  geom_col(width=0.5)+
  theme(axis.title.x=element_blank(),axis.text.x=element_blank())
fullback_graph<-(f1/f2/f3)|(f4/f5/f6)|(f7/f8/f9)

fullback_graph+ plot_layout(guides = 'collect')+
  plot_annotation(title = "Full back stats per 90mins")&
  theme(legend.position = "top")
```

* In comparison with other players, there is a similarity between what Marcelo did 2016/2017 and 2017/2018 and what Hakimi did so far, their numbers are close.
* There is a noticeable overall weakness in comparison with others players as his numbers are not so good. However, this can be explained by his younger age as a 23-year-old, this gives him time to improve his abilities.
* We notice that one of the qualities of Achraf Hakimi is that he is fast to develop and easy to adapt with a new team, manager and environment, and this is one of the things biggest assets.
